```python
import numpy as np
import sympy as sp
print(sp.__version__)
import itertools
N = 32
```

```python
dick = dict()
kcid = list()
for c in range(ord('А'), ord('А') + 32):
    # print(c - 1040, chr(c))
    dick[chr(c)] = c - 1040
    kcid.append(chr(c))
dick
# хайпраклинал хуй
word = "хайпраклинал"
word = [dick[c] for c in word.upper()]
word
```

```python
def matrix2word(matrix) -> str:
    matrix_word = str()
    for a in matrix:
        matrix_word += kcid[a]
    return matrix_word
```

```python
A = sp.Matrix(2, 2, [14, 15, 5, 16])    # ОПЕР 
print(matrix2word(A), matrix2word(A.inv_mod(N)), A.det() % N)

B = sp.Matrix(3, 3, [25, 0, 2, 5, 11, 28, 13, 0, 31]) # ЩАВЕЛЬНАЯ
print(matrix2word(B), matrix2word(B.inv_mod(N)), B.det() % N)

C = sp.Matrix(4, 4, [3, 8, 4, 16, 14, 12, 5, 21, 0, 13, 8, 7, 0, 22, 8, 31]) # ГИДРОМЕХАНИЗАЦИЯ
print(matrix2word(C), matrix2word(C.inv_mod(N)), C.det() % N)
```

```python
def fun(word, chunk_size):
    return [word[i:i + chunk_size] for i in range(0, len(word), chunk_size)]

def encryption(word, matrix):
    cipher = []
    for piece in fun(word, matrix.shape[0]):
        pos = (matrix * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        cipher += pos
    return matrix2word(cipher)

def transcript(cipher, matrix):
    transcript = []
    for piece in fun(cipher, 2):
        pos = (matrix.inv_mod(N) * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        transcript += pos
    return matrix2word(cipher)
encryption(word, A)
```

```python
# Преобразование python элементов в LaTeX

# Преобразование dick в несколько отдельных LaTeX-таблиц по 16 символов в каждой
row_len = 16
keys = list(dick.keys())
values = list(dick.values())

latex_tables = ""
for i in range(0, len(dick), row_len):
    header = " & ".join(str(k) for k in keys[i:i+row_len])
    vals = " & ".join(str(v) for v in values[i:i+row_len])
    table = "\\begin{tabular}{|" + "c|"*len(keys[i:i+row_len]) + "}\n\\hline\n"
    table += header + " \\\\\n"
    table += vals + " \\\\\n\\hline\n"
    table += "\\end{tabular}\n\n"
    latex_tables += table

print(latex_tables)

# Табличка: слово и числовые значения
word_str = ''.join([kcid[x] for x in word])
word_nums = ' & '.join(str(x) for x in word)
word_letters = ' & '.join(f"{kcid[x]}" for x in word)
latex_word_table = f"""
\\begin{{tabular}}{{|{'c|'*len(word)}}}
\\hline
{word_letters} \\\\
{word_nums} \\\\
\\hline
\\end{{tabular}}
"""
print(latex_word_table)

# представление матриц

latex_blocks = []
def matrix_to_latex_pmatrix(matrix, kcid):
    rows, cols = matrix.shape
    #Числовое
    num_rows = []
    for i in range(rows):
        num_rows.append(" & ".join(str(int(matrix[i, j])) for j in range(cols)))
    num_latex = " \\\\ ".join(num_rows)
    #Буквенное 
    let_rows = []
    for i in range(rows):
        let_rows.append(" & ".join(f"\\text{{{kcid[int(matrix[i, j])]}}}" for j in range(cols)))
    let_latex = " \\\\ ".join(let_rows)
    return num_latex, let_latex

matrices = {'A': A, 'B': B, 'C': C}
for name, mat in matrices.items():
    num_latex, let_latex = matrix_to_latex_pmatrix(mat, kcid)
    latex_blocks.append(f"\\[ {name} = \\begin{{pmatrix}}{num_latex}\\end{{pmatrix}} = \\begin{{pmatrix}}{let_latex}\\end{{pmatrix}} \\]")

print("\n\n".join(latex_blocks))
```
