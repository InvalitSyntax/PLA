```python
def fun(word, chunk_size):
    return [word[i:i + chunk_size] for i in range(0, len(word), chunk_size)]

def encryption(word, matrix):
    cipher = []
    for piece in fun(word, matrix.shape[0]):
        pos = (matrix * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        cipher += pos
    return cipher

def transcript(cipher, matrix):
    transcript = []
    for piece in fun(cipher, matrix.shape[0]):
        pos = (matrix.inv_mod(N) * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        transcript += pos
    return transcript
```

```python
def printtimest(matrix, vector, result):
    # Формируем LaTeX для вектора
        vector_latex = r"\begin{pmatrix} " + r" \\ ".join([str(x) for x in vector]) + r" \end{pmatrix}"
        
        # Формируем LaTeX для матрицы
        matrix_rows = []
        for i in range(matrix.shape[0]):
            row = " & ".join([str(x) for x in matrix.tolist()[i]])
            matrix_rows.append(row)
        matrix_latex = r"\begin{pmatrix} " + r" \\ ".join(matrix_rows) + r" \end{pmatrix}"
        
        # Формируем LaTeX для результата
        result_latex = r"\begin{pmatrix} " + r" \\ ".join([str(x) for x in result]) + r" \end{pmatrix}"
        
        # Выводим полное уравнение (матрица умножается на вектор-столбец)
        print('\\[' + matrix_latex + r" \cdot " + vector_latex + r" = " + result_latex + r" \pmod{32}" + "\\]")
        
def encryption_latex(word, matrix):
    cipher = []
    for piece in fun(word, matrix.shape[0]):
        pos = (matrix * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        printtimest(matrix, piece, pos)
        cipher += pos
    return cipher

def transcript_latex(cipher, matrix):
    transcript = []
    for piece in fun(cipher, matrix.shape[0]):
        pos = (matrix.inv_mod(N) * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        printtimest(matrix.inv_mod(N), piece, pos)
        transcript += pos
    return transcript
```

```python
def word_to_latex_table(word, numbers):
    """
    Возвращает LaTeX-код для таблички соответствия букв и числовых значений, по центру.
    """
    header = " & ".join(list(word.upper()))
    nums = " & ".join(str(n) for n in numbers)
    latex_code = (
        "\\begin{center}\n"
        "\\begin{tabular}{|" + "c|" * len(word) + "}\n\\hline\n"
        f"{header} \\\\\n\\hline\n"
        f"{nums} \\\\\n\\hline\n"
        "\\end{tabular}\n"
        "\\end{center}"
    )
    return latex_code

def word_to_latex_5bin_table(word, numbers):
    """
    Возвращает LaTeX-код для таблички соответствия букв и числовых значений, по центру.
    """
    header = " & ".join(list(word.upper()))
    nums = " & ".join(f"{n:05b}" for n in numbers)
    latex_code = (
        "\\begin{center}\n"
        "\\begin{tabular}{|" + "c|" * len(word) + "}\n\\hline\n"
        f"{header} \\\\\n\\hline\n"
        f"{nums} \\\\\n\\hline\n"
        "\\end{tabular}\n"
        "\\end{center}"
    )
    return latex_code

def array_to_latex_pmatrix_by_columns(arr, n):
    """
    Преобразует массив arr в матрицу высотой n по столбцам и возвращает LaTeX-код для pmatrix по центру.
    Например: arr=[1,2,3,4], n=2 -> \begin{pmatrix} 1 & 3 \\ 2 & 4 \end{pmatrix}
    """
    cols = len(arr) // n
    matrix = [[0 for _ in range(cols)] for _ in range(n)]
    for idx, val in enumerate(arr):
        row = idx % n
        col = idx // n
        matrix[row][col] = val
    latex_code = "\\[\n\\begin{pmatrix}\n"
    for i, row in enumerate(matrix):
        latex_code += " & ".join(str(x) for x in row)
        if i != n - 1:
            latex_code += " \\\\\n"
    latex_code += "\n\\end{pmatrix}\n\\]"
    return latex_code

import re

def parse_latex_matrix(latex_str):
    """
    Парсит LaTeX-матрицу и возвращает список списков (матрицу).
    """
    # Удаляем все \begin{pmatrix}, \end{pmatrix}, пробелы и \left, \right
    latex_str = re.sub(r'\\(begin|end)\{pmatrix\}', '', latex_str)
    latex_str = latex_str.replace('\\left', '').replace('\\right', '')
    latex_str = latex_str.strip()
    # Разбиваем на строки
    rows = [row.strip() for row in latex_str.split('\\\\') if row.strip()]
    matrix = []
    for row in rows:
        matrix.append([elem.strip() for elem in row.split('&')])
    return matrix

def latex_matrix_mult_expr(A_latex, B_latex):
    """
    Принимает две latex-матрицы (строкой), возвращает latex-матрицу произведения с расписанными выражениями.
    """
    A = parse_latex_matrix(A_latex)
    B = parse_latex_matrix(B_latex)
    n, m = len(A), len(A[0])
    m2, k = len(B), len(B[0])
    assert m == m2, "Матрицы нельзя перемножить: несовпадает число столбцов и строк"
    # Считаем выражения
    result = []
    for i in range(n):
        row = []
        for j in range(k):
            expr = " + ".join([f"{A[i][s]}\\cdot{B[s][j]}" for s in range(m)])
            row.append(expr)
        result.append(row)
    # Формируем latex
    latex_code = "\\begin{pmatrix}\n"
    for i, row in enumerate(result):
        latex_code += " & ".join(row)
        if i != n - 1:
            latex_code += " \\\\\n"
    latex_code += "\n\\end{pmatrix}"
    return latex_code

def parse_latex_matrix_to_numbers(latex_str):
    """
    Парсит LaTeX-матрицу и возвращает список списков (матрицу) с числами.
    """
    latex_str = re.sub(r'\\(begin|end)\{pmatrix\}', '', latex_str)
    latex_str = latex_str.replace('\\left', '').replace('\\right', '')
    latex_str = latex_str.strip()
    rows = [row.strip() for row in latex_str.split('\\\\') if row.strip()]
    matrix = []
    for row in rows:
        matrix.append([int(elem.strip()) for elem in row.split('&')])
    return matrix

def latex_matrix_mult_numeric(A_latex, B_latex):
    """
    Перемножает две latex-матрицы (строкой) и возвращает latex-матрицу результата.
    """
    A = parse_latex_matrix_to_numbers(A_latex)
    B = parse_latex_matrix_to_numbers(B_latex)
    A_sp = sp.Matrix(A)
    B_sp = sp.Matrix(B)
    result = A_sp * B_sp
    return sp.latex(result)

# Пример использования:
A = r"""\begin{pmatrix}21 & 16\\5 & 13\end{pmatrix}"""
B = r"""\begin{pmatrix}
21 & 9 & 16 & 10 & 8 & 0 \\
0 & 15 & 0 & 11 & 13 & 11
\end{pmatrix}"""

print(latex_matrix_mult_expr(A, B))
```

```python
# Порождающая матрица G (4×7) для кода Хэмминга (7,4)
G = np.array([
    [1, 1, 0, 1],  # P1: контролирует D1, D2, D4
    [1, 0, 1, 1],  # P2: контролирует D1, D3, D4  
    [1, 0, 0, 0],  # D1
    [0, 1, 1, 1],  # P3: контролирует D2, D3, D4
    [0, 1, 0, 0],  # D2
    [0, 0, 1, 0],  # D3
    [0, 0, 0, 1]   # D4
]).T  # Транспонируем, чтобы получить правильную размерность 4×7
print(G)

# Проверочная матрица H (3×7)
H = np.array([
    [1, 0, 1, 0, 1, 0, 1],  # P1 + D1 + D2 + D4
    [0, 1, 1, 0, 0, 1, 1],  # P2 + D1 + D3 + D4
    [0, 0, 0, 1, 1, 1, 1]   # P3 + D2 + D3 + D4
])

def binarray2char(arr):
    """Преобразование массива из 0 и 1 в символ"""
    return chr(int("".join(str(x) for x in arr), 2) + ord('А'))

def hamming_encode(data):
    """Кодирование данных"""
    N = len(data)
    M = len(data) * 7 // 4
    encoded = np.zeros(M, dtype=int)
    for i, j in zip(range(0, N, 4), range(0, M, 7)):
        block = data[i:i+4]
        x = np.dot(block, G) % 2
        encoded[j:j+7] = x
    return encoded

def hamming_decode(received):
    """Декодирование с исправлением ошибок"""
    M = len(received)
    N = len(received) * 4 // 7
    decoded = np.zeros(N, dtype=int)
    
    for index in range(0, M, 7):
        block = received[index:index+7]
        if len(block) < 7:
            block += [0] * (7 - len(block))  # Дополнение нулями до длины 7
        # Вычисление синдромов
        s = np.dot(H, block) % 2
        s1, s2, s3 = s
        # print(f"Received block #{index // 7}: {block}, Syndrome: {s}")
        # Определение позиции ошибки
        error_position = s1 * 1 + s2 * 2 + s3 * 4
        if error_position:
            print(f"\033[91mError detected in block {index + error_position-1}#{index // 7} {block} at position {error_position}\033[0m")
            # Исправление ошибки
            block[error_position - 1] ^= 1
        # print(f"Corrected block: {block} \t {block[[2,4,5,6]]}")
        decoded[index // 7 * 4:(index // 7 + 1) * 4] = block[[2,4,5,6]]
    print(f"Decoded array: {decoded} \t {''.join(binarray2char(decoded[i:i+5]) for i in range(0, len(decoded), 5))}")
    return decoded
```

```python
import random
from IPython.display import display
from IPython.display import Latex


def latexx(str):
    return latex(str, mat_str='pmatrix').replace(r'\left[', '').replace(r'\right]', '')

# Функции для перевода и тп
def encode_74(code, printing=True):
    oo = sp.Matrix(code).T
    out = (G * oo) % 2
    if printing:
        display(Latex(f"{latexx(G)}{latexx(oo)}={latexx(out)}"))
        print(f"{latexx(G)}{latexx(oo)}={latexx(out)}")
    return out

def decode_74(matrix_encoded, printing=True):
    out = (R * matrix_encoded) % 2
    outlist = out.T.tolist()
    if printing:
        display(Latex(f"{latexx(R)}{latexx(matrix_encoded)}={latexx(out)}"))
        print(f"{latexx(R)}{latexx(matrix_encoded)}={latexx(out)}")

    return outlist
        

def _flip_bits_in_matrix(matrix, num_bits, printing=True):
    m, n = matrix.shape
    total = m * n
    indices = random.sample(range(total), num_bits)
    matrix_err = matrix.copy()
    for idx in indices:
        row = idx // n
        col = idx % n
        old_val = matrix_err[row, col]
        matrix_err[row, col] = 1 - old_val
        if printing:
            print(f"Инвертирован бит в ячейке (строка={row+1}, столбец={col+1}), было={old_val}, стало={matrix_err[row, col]}")
    if printing:
        display(Latex(f"{latexx(matrix_err)} {row+1}\ {col+1}"))
        print(f"{latexx(R)}{latexx(matrix_err)}")
    return matrix_err

def flip_bits_in_matrix(matrix, num_bits, printing=True):
    m, n = matrix.shape
    matrix_err = matrix.copy()
    for col in range(n):
        rows_to_flip = random.sample(range(m), min(num_bits, m))
        for row in rows_to_flip:
            old_val = matrix_err[row, col]
            matrix_err[row, col] = 1 - old_val
            if printing:
                print(f"Инвертирован бит в ячейке (строка={row+1}, столбец={col+1}), было={old_val}, стало={matrix_err[row, col]}")
    if printing:
        display(Latex(f"{latexx(matrix_err)}"))
        print(f"{latexx(matrix_err)}")
    return matrix_err

def correcting(matrix, printing=True):
    symptom = (H * matrix) % 2
    vec = symptom.T.tolist()
    #print(vec)
    vec = [int(("".join(map(str, i)))[::-1], 2) for i in vec] # делаем реверс потому что
    #print(vec)
    vec_matrix = sp.Matrix(vec).T

    vec_add_matrix = []
    for i in vec:
        pp = [0, 0, 0, 0, 0, 0, 0]
        if i != 0:
            pp[i-1] = 1 - pp[i-1]
        vec_add_matrix.append(pp)
    
    vec_add_matrix = sp.Matrix(vec_add_matrix).T
    matrix_fixed = (vec_add_matrix + matrix) % 2

    if printing:
        display(Latex(f"{latexx(H)}{latexx(matrix)}={latexx(symptom)} = {latexx(vec_matrix)}"))
        print(f"{latexx(H)}{latexx(matrix)}={latexx(symptom)} = {latexx(vec_matrix)}")

        display(Latex(f"{latexx(vec_add_matrix)} + {latexx(matrix)} = {latexx(matrix_fixed)}"))
        print(f"{latexx(vec_add_matrix)} + {latexx(matrix)} = {latexx(matrix_fixed)}")

    return matrix_fixed


encoded_matrix = encode_74(code, printing=False)
broken_matrix = flip_bits_in_matrix(encoded_matrix, 1, printing=True)
corrected_matrix = correcting(broken_matrix, printing=True)
decoded_bin = decode_74(corrected_matrix)
decoded_word = bits_to_word(decoded_bin)
print(decoded_word)
print()
print(word_bin_latex_table(decoded_word, alphabet_bin))
```
