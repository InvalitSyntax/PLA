```python
import numpy as np
import sympy as sp
import itertools
N = 32
```

```python
dick = dict()
kcid = list()
for c in range(ord('А'), ord('А') + 32):
    # print(c - 1040, chr(c))
    dick[chr(c)] = c - 1040
    kcid.append(chr(c))
dick
# Ересеучитель
word = "хайпраклинал"
word = [dick[c] for c in word.upper()]
word
```

```python
def matrix2word(matrix) -> str:
    matrix_word = str()
    for a in matrix:
        matrix_word += kcid[a]
    return matrix_word
```

```python
A = sp.Matrix(2, 2, [14, 15, 5, 16])    # ОПЕР 
print(matrix2word(A), matrix2word(A.inv_mod(N)), A.det() % N)

B = sp.Matrix(3, 3, [25, 0, 2, 5, 11, 28, 13, 0, 31]) # ЩАВЕЛЬНАЯ
print(matrix2word(B), matrix2word(B.inv_mod(N)), B.det() % N)

C = sp.Matrix(4, 4, [3, 8, 4, 16, 14, 12, 5, 21, 0, 13, 8, 7, 0, 22, 8, 31]) # ГИДРОМЕХАНИЗАЦИЯ
print(matrix2word(C), matrix2word(C.inv_mod(N)), C.det() % N)
```

```python
def fun(word, chunk_size):
    return [word[i:i + chunk_size] for i in range(0, len(word), chunk_size)]

def encryption(word, matrix):
    cipher = []
    for piece in fun(word, matrix.shape[0]):
        pos = (matrix * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        cipher += pos
    return cipher

def transcript(cipher, matrix):
    transcript = []
    for piece in fun(cipher, matrix.shape[0]):
        pos = (matrix.inv_mod(N) * sp.Matrix(piece)).applyfunc(lambda x: x % 32).T.tolist()[0]
        transcript += pos
    return transcript
```
